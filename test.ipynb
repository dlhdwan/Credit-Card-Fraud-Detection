```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Thit lp d liu da trn cc phng trnh trong hnh
# To dy s lng Q t 0 n 40
q = np.linspace(0, 40, 100)

# nh ngha cc hm s (P)
# Demand (ng Cu): P = 40 - Q
price_demand = 40 - q

# MPC (Marginal Private Cost - Cung T nhn): P = 10 + Q
price_mpc = 10 + q

# MSC (Marginal Social Cost - Cung X hi): P = 20 + Q (tc l MPC + Tax)
price_msc = 20 + q

# 2. V  th
plt.figure(figsize=(8, 6)) # Kch thc nh

# V cc ng
plt.plot(q, price_demand, label='Demand (D)', color='navy', linewidth=2.5)
plt.plot(q, price_mpc, label='MPC (Supply)', color='green', linewidth=2.5)
plt.plot(q, price_msc, label='MSC (Social Cost)', color='red', linestyle='--', linewidth=2.5)

# 3. Trang tr  th ging hnh mu
plt.title('Market Analysis: Externalities (Phn tch Ngoi tc)', fontsize=14, pad=20)
plt.xlabel('Q (Quantity)', fontsize=12, loc='right')
plt.ylabel('P (Price)', fontsize=12, loc='top', rotation=0)

# Thit lp gii hn trc  ging hnh mu (P max = 50, Q max = 35)
plt.xlim(0, 35)
plt.ylim(0, 50)

# Bt li (Grid)  vung m
plt.grid(which='major', color='#666666', linestyle='-', alpha=0.6)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle='-', alpha=0.2)

# nh du cc im giao nhau (im cn bng)
# im cn bng th trng (Market Equilibrium): 10 + Q = 40 - Q => 2Q = 30 => Q=15, P=25
plt.plot(15, 25, 'ko') 
plt.text(15.5, 25.5, '$E_{market}$', fontsize=10)

# im ti u x hi (Social Optimum): 20 + Q = 40 - Q => 2Q = 20 => Q=10, P=30
plt.plot(10, 30, 'ro')
plt.text(10.5, 30.5, '$E_{social}$', fontsize=10, color='red')

# Thm nhn trc tip ln ng (ging hnh v tay)
plt.text(28, 8, 'D', fontsize=14, color='navy', fontweight='bold')
plt.text(28, 40, 'MPC', fontsize=14, color='green', fontweight='bold')
plt.text(22, 45, 'MSC', fontsize=14, color='red', fontweight='bold')

# Hin th
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Thit lp d liu
q = np.linspace(0, 40, 100)

# Cc phng trnh
price_demand = 40 - q          # Cu (D)
price_mpc = 10 + q             # Cung t nhn (MPC)
price_msc = 20 + q             # Cung x hi (MSC)

# 2. V khung  th
plt.figure(figsize=(8, 6))

# V cc ng
plt.plot(q, price_demand, label='Demand (D)', color='navy', linewidth=2)
plt.plot(q, price_mpc, label='MPC', color='green', linewidth=2)
plt.plot(q, price_msc, label='MSC', color='red', linestyle='--', linewidth=2)

# 3. im cn bng & Ti u ( tnh ton t h phng trnh)
# Market Equilibrium: Q=15, P=25
# Social Optimum: Q=10, P=30

# V im chm trn
plt.plot(15, 25, 'ko', markersize=8) # im en
plt.plot(10, 30, 'ro', markersize=8) # im 

# --- PHN IU CHNH V TR HIN TH ---

# Label cho E_market
# Dch chuyn y ln +2 n v, cn gia (ha='center')
plt.text(15, 25 + 2.5, '$E_{market}$', fontsize=11, ha='center', 
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=1))

# Label cho E_social
# Dch chuyn y ln +2.5 n v, cn gia, mu ch 
plt.text(10, 30 + 2.5, '$E_{social}$', fontsize=11, color='red', ha='center',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=1))

# ---------------------------------------

# 4. Trang tr hon thin
plt.title('Before Tax (Market Failure)', fontsize=14, pad=20)
plt.xlabel('Quantity (Q)', fontsize=12, loc='right')
plt.ylabel('Price (P)', fontsize=12, loc='top', rotation=0)

# Gii hn trc
plt.xlim(0, 35)
plt.ylim(0, 50)

# Li (Grid)
plt.grid(which='major', color='#666666', linestyle='-', alpha=0.4)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle=':', alpha=0.2)

# Nhn tn ng
plt.text(29, 8, 'D', fontsize=12, color='navy', fontweight='bold', bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))
plt.text(28, 42, 'MPC', fontsize=12, color='green', fontweight='bold', bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))
plt.text(21, 45, 'MSC', fontsize=12, color='red', fontweight='bold', bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Cu (D) v Cung (MPC c & MSC sau thu)
q = np.linspace(0, 40, 100)
price_demand = 40 - q          # Cu (D)
price_mpc = 10 + q             # Cung t nhn (MPC) - before tax (c)
price_msc = 20 + q             # Cung x hi (MSC) - ng cung mi sau thu (after tax)

# 2. V khung  th
plt.figure(figsize=(8, 6))

# V cc ng: Demand, MPC (c), MSC (sau thu)
plt.plot(q, price_demand, label='Demand (D)', color='navy', linewidth=2)
plt.plot(q, price_mpc, label='MPC (before tax)', color='green', linestyle='-', linewidth=2)
plt.plot(q, price_msc, label='MSC (after tax)', color='red', linewidth=2)

# 3. im cn bng & Ti u
# Market Equilibrium: Q=15, P=25
# Social Optimum: Q=10, P=30

plt.plot(10, 30, 'ro', markersize=6)  # Social optimum


# 3. im cn bng & Ti u ( tnh ton t h phng trnh)
# Market Equilibrium: Q=15, P=25
# Social Optimum: Q=10, P=30

# V im chm trn

plt.plot(10, 30, 'ro', markersize=8) # im 

# --- PHN IU CHNH V TR HIN TH ---

# Label cho E_social
# Dch chuyn y ln +2.5 n v, cn gia, mu ch 
plt.text(10, 30 + 2.5, '$E_{social}$', fontsize=11, color='red', ha='center',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=1))
# ---------------------------------------

# Exact red arrow at Q=20: from MPC (10+Q) up to MSC (20+Q)
q_point = 20
mpc_at_q = 10 + q_point
msc_at_q = 20 + q_point
plt.annotate(
    '',
    xy=(q_point, msc_at_q),      # arrow head at MSC
    xytext=(q_point, mpc_at_q),  # arrow tail at MPC
    arrowprops=dict(arrowstyle='-|>', color='red', linewidth=3, shrinkA=0, shrinkB=0, mutation_scale=20),
    annotation_clip=False
)
# Label to the right of the arrow, vertically centered
plt.text(q_point + 0.5, (mpc_at_q + msc_at_q) / 2.4, 'External benefit', color='red', fontsize=12, va='center', ha='left')

# Thm nhn trc tip cho cc ng (ging hnh v tay)
plt.text(29, 8, 'D', fontsize=12, color='navy', fontweight='bold', bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))
plt.text(28, 42, 'MPC (before tax)', fontsize=11, color='green', fontweight='bold', bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))
plt.text(15, 45, 'MSC <=> MPC (after tax)', fontsize=12, color='red', fontweight='bold', bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))

# 4. Trang tr hon thin
plt.title('After Tax (Correction)', fontsize=14, pad=20)
plt.xlabel('Quantity (Q)', fontsize=12, loc='right')
plt.ylabel('Price (P)', fontsize=12, loc='top', rotation=0)
plt.xlim(0, 35)
plt.ylim(0, 50)
plt.grid(which='major', color='#666666', linestyle='-', alpha=0.4)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle=':', alpha=0.2)

plt.tight_layout()
plt.show()
```
