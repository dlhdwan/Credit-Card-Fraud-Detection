```python
#Vit hm Cut_off chun IQR
def outlier_cutoff(data, feature_name):
    data_fraud = data[feature_name].loc[data['Class'] == 1].values
    q1,q3 = np.percentile(data_fraud, 25), np.percentile(data_fraud, 75)
    iqr = q3-q1
    
    cut_off = iqr*1.5
    lower = q1 - cut_off
    upper = q3 + cut_off
    
    outlier_remove = data[ (data['Class'] == 1) & ( (data[feature_name] < lower) | (data[feature_name] > upper) )]
    

    print(f'Feature {feature_name}:')
    print(f' - IQR: {iqr}')
    print(f' - Lower Bound: {lower} | Upper Bound: {upper}')
    print(f'S outlier s b loi b trong feature {feature_name}: {len(outlier_remove)}')
    
    return data.drop(outlier_remove.index)
    
    
```

```python
def remove_outlier_from_train(X_df, y, col_name):
    # Ly index ca cc dng l Fraud (Class 1)
    no_fraud_indices = np.where(y == 0)[0]
    
    # Ly gi tr ca ct cn xt ti cc dng Fraud
    no_fraud_values = X_df.iloc[no_fraud_indices][col_name].values
    
    # Tnh IQR
    q1, q3 = np.percentile(no_fraud_values, 25), np.percentile(no_fraud_values, 75)
    iqr = q3 - q1
    cut_off = iqr * 1.5
    lower, upper = q1 - cut_off, q3 + cut_off
    
    # Tm cc index l outlier (L Fraud V nm ngoi vng an ton)
    # Lu : no_fraud_indices l index trong y, cn map ng
    outliers_mask = (no_fraud_values < lower) | (no_fraud_values > upper)
    outliers_indices = no_fraud_indices[outliers_mask]
    
    if len(outliers_indices) > 0:
        print(f"Feature {col_name}:  loi b {len(outliers_indices)} outliers t lp No Fraud.")
        
        # Xa dng khi X v y
        X_df = X_df.drop(index=outliers_indices).reset_index(drop=True)
        y = np.delete(y, outliers_indices)
    else:
        print(f"Feature {col_name}: Khng tm thy outlier no trong lp No Fraud.")
        
    return X_df, y
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report, roc_curve, auc
```

```python
def plot_model_performance(model, X_test, y_test, y_pred, model_name):
    # 1. In Classification Report
    print(f"--- {model_name} Report ---")
    print(classification_report(y_test, y_pred))
    
    # 2. V Confusion Matrix v ROC Curve trn cng 1 hng
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax[0], annot_kws={"size": 16})
    ax[0].set_title(f'{model_name} - Confusion Matrix', fontsize=14)
    ax[0].set_xlabel('Predicted Label')
    ax[0].set_ylabel('True Label')
    
    # ROC Curve
    # Kim tra xem model c h tr predict_proba khng
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_test, y_prob)
        roc_auc = auc(fpr, tpr)
        
        ax[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
        ax[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        ax[1].set_xlim([0.0, 1.0])
        ax[1].set_ylim([0.0, 1.05])
        ax[1].set_xlabel('False Positive Rate')
        ax[1].set_ylabel('True Positive Rate')
        ax[1].set_title(f'{model_name} - ROC Curve', fontsize=14)
        ax[1].legend(loc="lower right")
    else:
        ax[1].text(0.5, 0.5, "Model does not support proba", ha='center')
    
    plt.tight_layout()
    plt.show()
    print("-" * 60)
```

```python
# ...existing code...
from sklearn.metrics import precision_recall_curve

def test_thresholds(model, X_test, y_test):
    # 1. Ly xc sut d on (Probability) thay v nhn 0/1
    # Ct [:, 1] l xc sut thuc lp Fraud
    y_prob = model.predict_proba(X_test)[:, 1]
    
    print(f"{'Threshold':<10} | {'Precision':<10} | {'Recall':<10} | {'F1-Score':<10} | {'Fraud Found':<12}")
    print("-" * 65)
    
    # 2. Th cc ngng t 0.5 n 0.99
    thresholds_to_test = [0.3,0.4, 0.5, 0.6, 0.7, 0.8, 0.85, 0.9, 0.95, 0.98, 0.99, 0.999]
    
    best_f1 = 0
    best_threshold = 0.5
    
    for t in thresholds_to_test:
        # Chuyn xc sut thnh nhn da trn ngng t
        y_pred_t = (y_prob >= t).astype(int)
        
        # Tnh cc ch s
        prec = precision_score(y_test, y_pred_t, zero_division=0)
        rec = recall_score(y_test, y_pred_t)
        f1 = f1_score(y_test, y_pred_t)
        fraud_found = np.sum((y_pred_t == 1) & (y_test == 1)) # S ca Fraud bt ng
        
        print(f"{t:<10} | {prec:.4f}     | {rec:.4f}     | {f1:.4f}     | {fraud_found}/{np.sum(y_test)}")

        if f1 > best_f1:
            best_f1 = f1
            best_threshold = t

    print("-" * 65)
    print(f"Ngng c F1-Score tt nht theo tnh ton: {best_threshold}")
    return y_prob
```
